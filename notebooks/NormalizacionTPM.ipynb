```python
gff_g = pd.DataFrame(gff[8])
gff_g = gff_g[8].str.split(';', expand=True)
gff_g
```

```python
gff_g[0] = gff_g[0].astype(str)
gff_g[1] = gff_g[1].astype(str)
```

```python
import re

gen_parent = pd.DataFrame()

# Usando el DataFrame 'gff_g' que ya tienes
gen_parent['gen'] = gff_g[0].apply(lambda x: re.search(r'ID=[^:]+:(.+)', x).group(1) if ':' in x else re.search(r'ID=([^:]+)', x).group(1) if 'ID=' in x else x)
gen_parent['parent'] = gff_g[1].apply(lambda x: re.search(r'Parent=([^:]+)', x).group(1) if 'Parent' in x else None)
```

```python
gen_parent.iloc[:, 0] = gen_parent.iloc[:, 0].str.split('.', expand=True)[0]
gen_parent.iloc[:,1] = gen_parent.iloc[:, 1].str.split('.', expand=True)[0]
gen_parent
```

```python
len(intvel_df.index.intersection(gen_parent['parent']))
```

```python
transcript = gen_parent[gen_parent.iloc[:, 1].isin(intvel_df.index)]
transcript.drop_duplicates()
```

```python
len(transcript['parent'].unique())
```

```python
transcript_knowe = transcript[transcript.iloc[:, 0].isin(knowe['gen'])]
transcript_knowe
```

```python
len(transcript_knowe['parent'].unique())
#-63
```

```python
merged_trans = pd.merge(knowe[['id', 'gen']], transcript_knowe[['gen','parent']], on='gen')
merged_trans = merged_trans.drop_duplicates()
merged_trans
```

```python
merged_trans_l = pd.merge(knowc[['chr','l1','l2','rid','id', 'id2']],merged_trans[['id','gen','parent']], on='id')
merged_trans_l = merged_trans_l.drop_duplicates()
```

```python
#-23
len(merged_trans_l['parent'].unique())
```

```python
unique_parents = merged_trans_l['parent'].value_counts().where(lambda x: x == 1).dropna().index.tolist()
```

```python
len(unique_parents)
#-103
```

```python
duplicated_parents = merged_trans_l['parent'].value_counts().where(lambda x: x == 2).dropna().index.tolist()
len(duplicated_parents)
```

```python
merged_trans_l= merged_trans_l.loc[merged_trans_l['parent'].isin(unique_parents)]
```

```python
merged_trans_l['l'] = merged_trans_l['l2'] - merged_trans_l['l1'] + 1
```

```python
merged_trans_l
```

```python
intvel_df
```

```python
intvel_df_filtered = intvel_df.loc[merged_trans_l['parent']]
```

```python
intvel_df_filtered
```

```python
merged_trans_l = merged_trans_l.set_index('parent')
intvel_df_filtered_nor = intvel_df_filtered.div(merged_trans_l['l'], axis=0)
intvel_df_filtered_nor
```

```python
sum_counts = intvel_df_filtered_nor.sum()
norm_factor = 1000000

intvel_df_tpm = intvel_df_filtered_nor.div(sum_counts, axis=1) * norm_factor
```
