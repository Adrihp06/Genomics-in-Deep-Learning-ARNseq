```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
DATA_PATH = '/home/adrihp06/AI_research/tensorflow/data/NNpro/data/'
idf = pd.read_csv(DATA_PATH + 'data_pro/intveld_dataprep_tpm_pro_01.csv', index_col=0)
idf_m = pd.read_csv(DATA_PATH + 'data_pro/meta_dataprep_sa_01.csv')
idf_ori = pd.read_csv(DATA_PATH + 'data_pro/intveld_dataprep.csv', index_col=0)
idf_corr = pd.read_csv(DATA_PATH + 'data_pro/intveld_dataprep_tpm_pro_02_corr.csv', index_col=0)
```

```python
idf
```

```python
clases = [
    'Asymptomatic Controls',
    'Non-Small-Cell Lung Cancer',
    'Ovarian Cancer',
    'Glioma',
    'Pancreatic Cancer',
    'Head and Neck Cancer',
    'Breast Cancer',
    'Colorectal Cancer',
    'Cholangiocarcinoma',
    'Melanoma',
    'Sarcoma'
]
```

```python
filas_seleccionadas = {}

for clase in clases:
    filas_clase = idf_m[idf_m['Type'] == clase].sample(n=50)
    filas_seleccionadas[clase] = filas_clase
```

```python
idf_ori
```

```python
pacientes_idf = pd.concat([filas_seleccionadas[clase] for clase in clases])
idf_nuevo = idf.T.loc[pacientes_idf['Patient']]

pacientes_idf_m = pd.concat([filas_seleccionadas[clase] for clase in clases])
idf_m_nuevo = idf_m.loc[pacientes_idf_m.index]
```

```python
pacientes_idf
```

```python
idf_nuevo = idf_nuevo.T
```

```python
idf_m_nuevo.drop('Type_encoded', axis=1)
```

```python
le = LabelEncoder()
le.fit(idf_m_nuevo['Type'])
pacientes_idf['Type_encoded'] = le.transform(idf_m_nuevo['Type'])
```

```python
mask = idf_m['Patient'].isin(idf_ori.T.index)
idf_m_filtrado = idf_m[mask]

filas_seleccionadas_2 = {}

for clase in clases:
    filas_clase = idf_m_filtrado[idf_m_filtrado['Type'] == clase].sample(n=50)
    filas_seleccionadas_2[clase] = filas_clase

pacientes_idf_ori = pd.concat([filas_seleccionadas_2[clase] for clase in clases])
idf_ori_nuevo = idf_ori.T.loc[pacientes_idf_ori['Patient']]
```

```python
idf_ori_nuevo = idf_ori_nuevo.T
```

```python
pacientes_idf_ori.drop('Type_encoded', axis=1)
le.fit(pacientes_idf_ori['Type'])
pacientes_idf_ori['Type_encoded'] = le.transform(pacientes_idf_ori['Type'])
```

```python
pacientes_idf_ori
```

```python
filas_seleccionadas_3 = {}

for clase in clases:
    filas_clase = idf_m[idf_m['Type'] == clase]
    filas_seleccionadas_3[clase] = filas_clase
```

```python
pacientes_idf_filtered = pd.concat([filas_seleccionadas_3[clase] for clase in clases])
idf_filtered = idf.T.loc[pacientes_idf_filtered['Patient']]
```

```python
pacientes_idf_filtered
```

```python
idf_filtered = idf_filtered.T
idf_filtered
```

```python
pacientes_idf_filtered.drop('Type_encoded', axis=1)
le.fit(pacientes_idf_filtered['Type'])
pacientes_idf_filtered['Type_encoded'] = le.transform(pacientes_idf_filtered['Type'])
```

```python
idf_corr
```

```python
idf_filtered_corr = idf_corr.T.loc[pacientes_idf_filtered['Patient']]
idf_filtered_corr = idf_filtered_corr.T
idf_filtered_corr
```

```python
pacientes_idf.to_csv(DATA_PATH + 'data_pro/meta_dataprep_sa_sample.csv', index=False)
idf_nuevo.to_csv(DATA_PATH + 'data_pro/intveld_dataprep_tpm_pro_sample.csv')

pacientes_idf_ori.to_csv(DATA_PATH + 'data_pro/meta_dataprep_sample.csv', index=False)
idf_ori_nuevo.to_csv(DATA_PATH + 'data_pro/intveld_dataprep_sample.csv')

pacientes_idf_filtered.to_csv(DATA_PATH + 'data_pro/meta_dataprep_filtered.csv', index=False)
idf_filtered.to_csv(DATA_PATH + 'data_pro/intveld_dataprep_filtered.csv')

idf_filtered_corr.to_csv(DATA_PATH + 'data_pro/intveld_dataprep_corr_02.csv')
```
